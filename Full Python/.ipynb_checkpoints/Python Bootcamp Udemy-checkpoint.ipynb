## D2. Tipe Data Dasar : Null, Boolean, Numeric, Text
-------

- Null Type: Tipe data null dalam Python digunakan untuk menyimpan nilai kosong atau tidak ada yang dinyatakan dengan None.
- Boolean Type: Tipe data boolean atau bool digunakan untuk menyimpan nilai kebenaran (True, False) dari suatu ekspresi logika.
- Numeric Type: Tipe data yang digunakan untuk menyimpan data berupa angka. Terdapat dua macam tipe data numeric, yaitu int untuk menyimpan bilangan bulat (e.g.: 0, 1, 2, 404, -500, -1000) dan float untuk menyimpan bilangan riil (e.g.: 0.5, 1.01, 2.05, 4.04)
- Text Type: Pada Python, tipe data string (str) digunakan untuk menyimpan data teks. Tipe data string dimulai dengan tanda kutip (baik kutip satu/ dua) dan diakhir dengan tanda kutip. Contoh: "Teks", "Contoh teks menggunakan Python", dan 'Teks pada Python'.

```python
#Deklarasi Variabel
tipe_null = ""
tipe_boolean = True
tipe_booleanX = False
tipe_boolean1 = 5 < 10
tipe_boolean2 = 5 > 10
tipe_int = 10
tipe_float = 10.0
tipe_str = "Sepuluh"

#Print nilai variabel
print (tipe_null) #output tidak menampilkan apa apa
print (tipe_boolean) #output menampilkan nilai kebenaran True atau 1
print (tipe_booleanX) #output menampilkan nilai kebenaran False atau 0
print (tipe_boolean1)
print (tipe_boolean2)
print (tipe_int)
print (tipe_float)
print (tipe_str)
```

## D3. Python Sebagai Dynamic Typing
------------

Python adalah bahasa pemrograman bertipe dinamis (kebalikan dari Java dan C/C++ yang bertipe statis) artinya programmer tidak perlu mendefinisikan tipe data variabel saat memprogram.


### D3.1. Dynamic Typing pada Python

Perhatikan contoh pendeklarasian variabel dalam bahasa Python berikut ini :

    variabel = 4
    print(variabel)
.

    Output :
    4

Kemudian kita ubah nilai variabel dengan angka 10.
    
    variabel = 4
    variabel = 10
    print(variabel)
.

    Output :
    10
Outputnya berubah menjadi 10 karena nilai variabel1 yang awalnya 4 telah diubah menjadi 10.

Artinya variabel bisa diubah nilainya dengan nilai bertipe data sama (dalam kasus di atas adalah integer). Lalu bagaimana jika ingin mengubahnya dengan nilai bertipe data string?

    variabel = 4
    variabel = 10
    variabel = "Hallo"
    print(variabel)
.

    Output :
    Hallo
Outputnya memunculkan 'Hello'. Artinya variabel bisa diganti dengan tipe data yang berbeda.

Hal ini membuktikan bahwa Python bersifat Dynamic Typing.

## D4 Casting
---------

Casting merupakan sebuah metode atau cara untuk menentukan tipe data variabel.

Walaupun Python merupakan bahasa yang menganut Dynamic Typing, namun bukan berarti tidak bisa menentukan tipe data dalam variabelnya. Penentuan tipe data variabel ini menggunakan casting.

Penggunaan casting pada Python dilakukan dalam fungsi konstruktor, karena dalam Python berlaku konsep "Everything in Python is an Object". Oleh karena itu Python menggunakan class untuk mendefinisikan tipe data termasuk tipe data primitifnya.

Note :

- Konstruktor adalah method untuk menginstansiasi objek (membuat objek). Tugasnya yaitu menginisialisasi (memberikan nilai) pada objek dari class tertentu.
- Class adalah cetak biru atau blueprint dari object. Class digunakan hanya untuk membuat kerangka dasar. Yang akan kita pakai nantinya adalah hasil cetakan dari class, yakni object.
- Object adalah hasil cetak dari class, atau hasil ‘konkrit’ dari class.

Contoh, jika menggunakan analogi class Laptop, maka objek dari class Laptop bisa berupa: laptop_andi, laptop_anto, laptop_zein, dan lain-lain. Objek dari class laptop akan memiliki karakteristik masing-masing yang mana dapat diinisialisasi melalui konstruktor.

    (Selebihnya tentang konstruktor, class, dan objek, akan dijelaskan di materi Object Oriented Programming Python)

```python
#Membuat variabel x yang diisi oleh angka 2
x = 2
```

```python
#Cek tipe data dari x
type(x)
```

```python
#Melakukan casting sementara
float(x)
```

```python
#Jika casting dilakukan seperti itu, maka ketika dicek tipe data dari x tidak akan berubah
#karena dia mengubah nilai x ke float hanya sementara
#artinya nilai x yang asli tetap integer yang awal dideklarasikan
type(x)
```

```python
#Melakukan casting secara total
x = float(2)

#maka variabel x akan menjadi float
```

```python
#jika dicek tipe datanya maka akan float, karena variabel x sudah dicasting secara default dengan float
type(x)
```

### D5.6. Indexing & Slicing

Seperti banyak bahasa pemrograman populer lainnya, string dalam Python menggunakan konsep array yang mewakili karakter unicode dalam index tertentu.

Konsep ini, dapat dikembangkan untuk memahami indexing dan slicing, yang mana memanfaatkan index pada string.

- Indexing adalah memanggil nilai karakter pada index tertentu. Misalnya kita kita ingin memanggil huruf yang terletak pada index ke 3 dalam string "Python", maka akan muncul output 'h' karena 'h' berada di index ke 3.

- Slicing adalah mengiris karakter pada string berdasarkan index. Misalnya kita ingin mengiris huruf 't' pada string "Python", maka kita harus mengiris index ke-2.

```python
bahasa = "Python"
#INDEXING (index dimulai dari 0)
print(bahasa[2])  #memanggil index ke 2 dari variabel bahasa 
print(bahasa[3]) #memanggil index ke 3 dari variabel bahasa

#SLICING
print(bahasa[1:4]) #mengiris index 1 sampai ke 4 (akan muncul karakter yang letaknya di index ke 1 sampai 4)
print(bahasa[2:5]) #mengiris index 2 sampai ke 5 dari variabel bahasa

#SLICING DARI ARAH KEBALIKAN
print(bahasa[-3]) #mengiris index ke -3 dimulai dari kanan (index 0 tetap P)
print(bahasa[1:-2])
```

### D5.7. String Formatting
Di sini kita akan mempelajari tentang cara-cara format string.

Ada tiga cara melakukan format pada string. Pertama dengan menggunakan fungsi format(), yang kedua dengan f"...", dan ketiga dengan menggunakan cara lama (menggunakan %).


##### Cara format()
Memformat string dengan fungsi format() dibuat dengan menggunakan tanda {} sebagai placeholder atau posisi substring yang akan digantikan. Kita biasa menggunakan argumen posisi atau kata kunci untuk menunjukkan urutan dari substring.

```python
# menggunakan posisi default 
default_order = "nama saya {}, teman saya {} dan {}".format("Budi", "Galih","Ratna") 
print('\n--- Urutan default ---') 
print(default_order)
```

```python
# menggunakan argument posisi 
positional_order = "nama saya {1}, teman saya {0} dan {2}".format("Budi", "Galih","Ratna") 
print('\n--- Urutan berdasarkan posisi ---') 
print(positional_order) 
```

Metode format() dapat memiliki spesifikasi format opsional. Misalnya, kita bisa menggunakan tanda < untuk rata kiri, > untuk rata kanan, ^ untuk rata tengah, dan sebagainya.

```python
#Format integer
print("{0} bila diubah jadi biner menjadi {0:b}".format(12))

#Format float
print("Format eksponensial: {0:e}".format(1566.345))

#Pembulatan
print("Sepertiga sama dengan: {0:.3f}".format(1/3))

#Meratakan string
print("|{:<10}|{:^10}|{:>10}|".format('beras', 'gula', 'garam'))
```

##### Cara f"..."
Hampir sama dengan format(), Memformat string dengan cara ini juga dibuat dengan menggunakan tanda {} sebagai placeholder atau posisi substring yang akan digantikan.

```python
nama = "Zein"
kampus = "UTY"
Jabatan = "CEO"

print(f"nama saya {nama}, saya lulusan {kampus}, jabatan saya {Jabatan}")
```

##### Cara %
Kita juga bisa menggunakan cara lama, yaitu dengan memanfaatkan operator % untuk melakukan format string.

```python
nama = 'Budi'
print('Nama saya %s' %nama)
```

### D6.1. List
Tipe data ini digunakan untuk menampung sekumpulan data secara terorganisir. List dan tuple merupakan bentuk dari tipe data sequence.

Tipe data list diawali dengan tanda kurung siku buka ( [ ), memisahkan setiap elemen di dalamnya dengan tanda koma ( , ) dan ditutup dengan kurung siku tutup ( ] ).

Contoh :

    contoh_list = [1, 'dua', 3, 4.0, 5 ]
    
Setiap elemen dari list memiliki indeks yang dimulai dari angka 0 dan terus bertambah satu nilainya hingga elemen terakhir dari list.

```python
contoh_list = [1, 'dua', 3, 4.0, 5]
print (contoh_list[0]) # Akan menampilkan angka 1 
print (contoh_list[1]) # Akan menampilkan string 'dua'
```

Tipe data list bersifat mutable yang berarti setiap elemen di dalam list dapat dirubah nilainya setelah proses pendeklarasian list.

```python
contoh_list = [1, 'dua', 3, 4.0, 5 ]
print(contoh_list[3]) # Akan menampilkan angka 4.0
contoh_list[3] = 'empat'

print(contoh_list[3]) # Akan menampilkan string "empat" karena telah diubah nilainya
```

### D6.2. Tuple
Tipe data tuple juga berfungsi untuk menampung sekumpulan data. Tipe data tuple diawali dengan tanda kurung buka ( ( ), memisahkan setiap elemen di dalamnya dengan tanda koma ( , ) dan ditutup dengan tanda kurung tutup ( ) ).

Contoh :

    contoh_tuple = ('Januari', 'Februari', 'Maret','April')

Aturan indeks dan cara mengakses elemen pada sebuah tuple serupa dengan list.

```python
contoh_tuple = ('Januari', 'Februari', 'Maret','April')
print(contoh_tuple[2]) # Akan menampilkan 'Maret'
```

Berbeda dengan tipe data list, tipe data tuple bersifat immutable yang berarti elemen pada tipe data tuple tidak dapat diubah setelah proses pendeklarasiannya. Ketika dirubah maka akan error.

```python
contoh_tuple = ('Januari', 'Februari', 'Maret','April')
contoh_tuple[2] = 'Desember'
```

## D7. Tipe data Sets
--------

### D7.1. Set
Serupa dengan tipe data sequence, tipe data set digunakan untuk menampung sekumpulan data dengan tipe lainnya. Terdapat dua jenis dari tipe data set yaitu, set dan frozenset.

Tipe data set diawali dengan tanda kurung buka kurawal ( { ), memisahkan setiap elemen di dalamnya dengan tanda koma ( , ) dan ditutup dengan tanda kurung tutup ( } ). Namun, berbeda dengan tipe data sequence, seperti list, tipe data objek tidak mengizinkan adanya elemen dengan nilai yang sama dan tidak memperdulikan urutan dari elemen.

Contoh :

```python
#LIST
contoh_list = ['Dewi', 'Budi', 'Cici', 'Linda', 'Cici']
print(contoh_list)
```

```python
#SET
contoh_set = {'Dewi', 'Budi', 'Cici', 'Linda', 'Cici'}
print(contoh_set)
```

Dari kedua contoh output pada program, dapat terlihat:

- Berbeda dengan tipe data set, tipe data list memperdulikan urutan dari setiap elemen saat list dideklarasikan.
- Berbeda dengan list yang mengizinkan adanya duplikasi elemen, tipe data set tidak mengizinkan adanya elemen dengan nilai yang sama di dalamnya.

### D7.2. Fozenset
Bagaimana dengan tipe data frozenset? Tipe data frozenset sebenarnya hanya merupakan set yang bersifat immutable, yang artinya setiap elemen di dalam frozenset tidak dapat diubah setelah proses deklarasinya. Untuk membuat tipe data frozenset perhatikan contoh berikut :

```python
contoh_frozen_set = ({'Dewi', 'Budi', 'Cici', 'Linda', 'Cici'})
print(contoh_frozen_set)
```

## D8. Mapping Type : Dictionary
--------

Tipe data mapping dapat digunakan untuk memetakan sebuah nilai ke nilai lainnya. Dalam Python, tipe data mapping disebut dengan istilah dictionary. Tipe data dictionary dapat dideklarasikan dengan diawali oleh tanda kurung buka kurawal ( { ), memisahkan setiap elemen di dalamnya dengan tanda koma ( , ) dan ditutup dengan tanda kurung tutup ( } ). Setiap elemen pada tipe data dictionary dideklarasikan dengan format:

"kunci" : "nilai"

Hal inilah yang membedakan tipe data mapping dengan tipe data set. Untuk lebih memperjelas tipe data mapping, perhatikan contoh berikut

```python
person = {'nama': 'John Doe', 'pekerjaan': 'Programmer'}
```

Potongan kode di atas adalah contoh pendeklarasian dari tipe data dictionary. Kemudian, bagaimana dengan cara mengakses informasi dari tipe data ini? Perhatikan potongan kode berikut

```python
print(person['nama'])
print(person['pekerjaan'])
```

Selanjutnya, Product Manager meminta kita untuk menghitung total harga jual dengan potongan harga. Tak lupa menyertakan pajak sebesar 10% dari nilai jual.

Di sini kita diberi tips oleh Product Manager untuk menyelesaikan tugas ini:

1. Data yang dinyatakan ke dalam dictionary
2. Hitung harga masing-masing data setelah dikurangi diskon
3. Hitung harga total
4. Hitung harga kena pajak
5. Cetak total_harga + total_pajak

```python
# Data yang dinyatakan ke dalam dictionary
sepatu = {"nama": "Sepatu Niko", "harga": 150000, "diskon": 30000} 
baju = {"nama": "Baju Unikloh", "harga": 80000, "diskon": 8000} 
celana = {"nama": "Celana Lepis", "harga": 200000, "diskon": 60000}

# Hitunglah harga masing-masing data setelah dikurangi diskon
harga_sepatu = sepatu["harga"] - sepatu["diskon"] 
harga_baju = baju["harga"] - baju["diskon"]
harga_celana = celana["harga"] - celana["diskon"]

# Hitung harga total
total_harga = harga_sepatu + harga_baju + harga_celana

# Hitung harga kena pajak
total_pajak = total_harga * 0.1

# Cetak total_harga + total_pajak
print(total_harga + total_pajak)
```
